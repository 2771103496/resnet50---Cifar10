```python
import torch
import torchvision
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets, transforms
from torch.autograd import Variable
import os
import gzip
from torch.utils.data import DataLoader, Dataset
import numpy as np
import matplotlib.pyplot as plt
import copy
import torchvision.models.resnet

epochs = 80

def img_show(img):
    img = img / 2 + 0.5  # 将img的像素区间转换为[0,1]
    np_img = img.numpy()
    plt.imshow(np.transpose(np_img, (1,2,0)))   # 图片必须以（H,W,C）的numpy形式才能输出显示
    plt.show()


transform = {
    "train": transforms.Compose([transforms.RandomResizedCrop(224),
                                 transforms.RandomHorizontalFlip(),
                                 transforms.ToTensor(),
                                 transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])]),
    "val": transforms.Compose([transforms.Resize(256),
                               transforms.CenterCrop(224),
                               transforms.ToTensor(),
                               transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])])}


trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform["train"])
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True, num_workers=0)

# CIFAR10的训练集一共有50000张图片，每张图片大小为（32,32,3）

testset = torchvision.datasets.CIFAR10(root='./data', train=False,download=True, transform=transform["val"])
testloader = torch.utils.data.DataLoader(testset, batch_size=64, shuffle=False, num_workers=0)

classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')


net = torchvision.models.resnet50(pretrained=True)
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
for param in net.parameters():
    param.requires_grad = True
fc_inputs = net.fc.in_features
net.fc = nn.Sequential(
    nn.Linear(fc_inputs, 10)
)
for param in net.fc.parameters():
    param.requires_grad = True
net = net.to(device)
loss_function = nn.CrossEntropyLoss()
optimizer = optim.Adam(net.parameters(), lr=0.001)
for epoch in range(epochs):
    net.train()
    running_loss = 0
    for ids, data in enumerate(trainloader, 0):
        images, labels = data
        images = images.to(device)
        labels = labels.to(device)
        optimizer.zero_grad()
        outputs = net(images)
        _, predicted = torch.max(outputs.data, 1)
        loss = loss_function(outputs, labels)
        total = labels.size(0)
        correct = (predicted == labels).sum().item()
        acc = correct / total
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        if ids % 50 == 49:
            print("epoch [{}/{}],iter[{}],acc{:.4f}, loss{:.4f}".format(epoch + 1, epochs, ids + 1, acc, running_loss / 50.0))
            running_loss = 0
torch.save(net.state_dict(), 'resnet50.pt')
net.eval()
acc = 0
total = 0
correct = 0 
with torch.no_grad():
  for i,(images, labels) in enumerate(testloader, 0):
    images = images.to(device)
    labels = labels.to(device)
    output = net(images)
    _, pre = torch.max(output, dim=1)
    correct += (pre == labels).sum().item()
    total += labels.size(0)
acc = correct / total
print('the accuracy is %.3f %%' %(acc * 100))

```
